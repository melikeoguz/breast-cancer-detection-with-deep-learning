<h2>1. Giriş</h2>

Meme kanseri, kadınlar arasında ikinci ölüm nedenidir. Erken teşhis ve ardından uygun kanser tedavisi ölümcül riski azaltabilir. Tümörler iyi veya kötü huylu olabilir. İyi huylu tümörler kanser değildir. Meme kanseri kötü huylu bir tümördür ve göğüs hücrelerinde gelişir.

Tıp uzmanları bir hastalığı tespit ederken hatalar yapabilirler. Veri madenciliği ve makine öğrenimi gibi teknolojinin yardımı, teşhis doğruluğunu önemli ölçüde artırabilir. Yapay Sinir Ağları (YSA), akıllı meme kanseri teşhisinde yaygın olarak kullanılmaktadır.


<h2>2. Yapay Sinir Ağları (Artificial Neural Network (ANN))</h2>

Yapay sinir ağları insan beyni örnek alınarak, öğrenme sürecinin matematiksel olarak modellenmesi sonucu ortaya çıkmıştır. Beyindeki biyolojik sinir ağlarının yapısını, öğrenme, hatırlama ve genelleme kabiliyetlerini taklit eder.

<center><img src="https://i.hizliresim.com/35n47ff.jpeg"></center>

YSA, sınıflandırma ve regresyon problemleri gibi yaygın veri madenciliği görevleri, örüntü tanıma, tıbbi teşhis, makine öğrenimi ve yapay zeka gibi çeşitli alanlarda akıllı bir araç olarak yaygın bir şekilde kullanılmaktadır.

Yapay sinir ağları, birçok hücreden meydana gelir ve bu hücreler eş zamanlı çalışarak karmaşık işleri gerçekleştirir. Öğrenme kabiliyeti olduğundan farklı öğrenme algoritmalarıyla öğrenebilir.  YSA, görülmemiş çıktılar için sonuç (bilgi) üretir. Örüntü tanıma ve sınıflandırma yapma ve eksik örüntüleri tamamlayabilme özellikleri vardır.  Hata toleransına sahiptir. Eksik veya belirsiz bilgiyle çalışabilir. Hatalı durumlarda dereceli bozulma(graceful degradation) gösterir. Ek olarak paralel çalışabilmekte ve gerçek zamanlı bilgiyi işleyebilmektedir.

Yapay sinir ağları başlıca teşhis, sınıflandırma, tahmin, kontrol, veri ilişkilendirme, veri filtreleme, yorumlama gibi alanlarda kullanılmaktadır. Hangi problem için hangi ağın daha uygun olduğunu belirlerken ağların özellikleri ile problemlerin özelliklerini karşılaştırmak gerekir.

Meme kanserinin sınıflandırılması, araştırmacılar ve bilim adamları için büyük bir zorluk oluşturan tıbbi bir uygulamadır. Birçok araştırma, YSA'nın meme kanseri teşhisinde iyi bir doğruluk sağladığını gösterdi.

<p></p>
Bu sınıflandırma problemi için bir <b>Keras 2.0</b> modeli oluşturulmaktadır.

```python
import tensorflow as tf 
from tensorflow import keras
```

<p></p>
<b>Sequential</b>, Keras’ta bir model oluşturmanın en kolay yoludur. Katman ile bir model katman oluşturmanıza izin verir. <b>‘Dense’</b>, katman türüdür. Dense, çoğu durumda çalışan standart bir katman türüdür. Yoğun bir katmanda, önceki katmandaki tüm düğümler mevcut katmandaki düğümlere bağlanır.
<ul>
    <li>
        <b>Input Layer:</b> Yapay sinir ağına dış dünyadan girdilerin geldiği katmandır. Bu katmanda dış dünyadan gelecek giriş sayısı kadar nöron bulunmasına rağmen genelde girdiler herhangi bir işleme uğramadan alt katmanlara iletilmektedir.
    </li>
     <li>
         <b>Hidden Layer:</b> Giriş katmanından çıkan bilgiler bu katmana gelir. Ara katman sayısı ağdan ağa değişebilir. Bazı yapay sinir ağlarında ara katman bulunmadığı gibi bazı yapay sinir ağlarında ise birden fazla ara katman bulunmaktadır. Ara katmanlardaki nöron sayıları giriş ve çıkış sayısından bağımsızdır. Birden fazla ara katman olan ağlarda ara katmanların kendi aralarındaki nöron sayıları da farklı olabilir. Ara katmanların ve bu katmanlardaki nöronların sayısının artması hesaplama karmaşıklığını ve süresini arttırmasına rağmen yapay sinir ağının daha karmaşık problemlerin çözümünde de kullanılabilmesini sağlar. Hidden, ReLU [0, +∞) aralığında değer alıyor.
    </li>
    <li>
        <b>Output Layer:</b> Ara katmanlardan gelen bilgileri işleyerek ağın girdi katmanından gelen verilere karşılık olan çıktıları üreten katmandır. Bu katmanda üretilen çıktılar dış dünyaya gönderilir. Geri beslemeli ağlarda bu katmanda üretilen çıktı kullanılarak ağın yeni ağırlık değerleri hesaplanır. Output her zaman (0,1) aralığında değer üretir.
    </li>
</ul>

<p><b>Aktivasyon Fonksiyonu</b></p>
Bir nöronun çıktısını belirleyen matematiksel bir denklemdir. Bu işlemin  yapılmasının nedeni doğrusallıktan kaçıp sonucunun non-linear yani doğrusal olmayan bir yapıda olmasını sağlamaktır. Eğer lineer bir çıktı olsaydı yani üssel olarak tek dereceli bir polinom olsaydı lineer regresyon gibi doğrusal bir denklem olurdu ve karmaşık öğrenme işlemleri bu denklemle gerçekleştirilemezdi.Sonuç olarak aktivasyon fonksiyonu, karmaşık yani doğrusal olmayan paternlerin öğrenilmesinde önemlidir. Bu fonksiyonlara bir kaç örnek Sigmoid, ReLU ve türevleri, tanh, swish, softmax olarak sıralanabilir.

<p><b>ReLU Fonksiyonu:</b></p>

<center><img src="https://i.hizliresim.com/lpldz6r.jpg"></center>

Genellikle Convolutional Neural Network (CNN)’te ve ara katmanlarda çok sık kullanılan ReLU fonksiyonunun ana avantajı aynı anda tüm nöronları aktive etmemesidir. Yani bir nöron negatif değer üretirse, aktive edilmeyeceği anlamına gelir.

Tablodan da görüleceği üzere negatif değerler üreten nöronlar sıfır değerini alır. Bu durum, ReLU’nun Hiperbolik Tanjant ve Sigmoid fonksiyonundan daha verimli ve hızlı çalışmasını sağlar. Bu nedenle ReLU, çok katmanlı sinir ağlarında daha çok tercih edilir.

Fakat ReLU mükemmel değildir. Grafiğe bakıldığında, fonksiyonun türevinin negatif değerlerde sıfır olduğu görülür. Bu durum Geri Yayılım (Back Propagation) esnasında parametrelerin güncellenmeyeceği, yani öğrenme işleminin olmayacağı anlamına gelir. Bu probleme Dying ReLU adı verilir.

<p><b>Sigmoid Fonksiyonu:</b></p>

<center><img src="https://i.hizliresim.com/mtp5j6y.jpg"></center>

En çok kullanılan aktivasyon fonksiyonlarından birisidir. Doğrusal olmayan(non-linear) bir fonksiyondur. Karar vermeye yönelik olasılıksal bir yaklaşımdır ve değer aralığı [0,1] arasındadır. Yani çıktının hangi sınıfa ait olduğuna dair olasılıksal bir değer verir. Sigmoid fonksiyonu sürekli yani türevlenebilir bir fonksiyon olduğundan öğrenme işlemi gerçekleşir.

Fakat Sigmoid de mükemmel değildir. Çünkü türev değeri, tablodan da görebildiği üzere, uç noktalarda sıfıra yakınsar (Vanishing Gradient).Bu durum Back Propagation sırasında öğrenmenin durmasına sebebiyet verir. Bu sebeple çok katmanlı ağlarda kullanılması tavsiye edilmez.

Diğer aktivasyon fonksiyonlardan bazıları şunlardır: 
<ul>
    <li>Basamak(Step) Fonksiyonu</li>
    <li>Doğrusal (Linear) Fonksiyon</li>
    <li>Hiperbolik Tanjant Fonksiyonu</li>
    <li>Sızıntı (Leaky) ReLU Fonksiyonu</li>
    <li>Softmax Fonksiyonu</li>
</ul>
    


```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Input(shape=(D,)),
  tf.keras.layers.Dense(1, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
    
])
```

<p></p>
Modelin derlenmesi <b>model.compile()</b> fonksiyonu ile yapılır. Bu fonksiyon optimizer, kayıp(loss) ve metrik(metrics) olmak üzere üç parametre alır.

<ul>
    <li>
       <b>optimizer</b> öğrenme oranını kontrol eder. Optimizer olarak <b>“adam”</b> kullanılmaktadır. Adam genellikle birçok durumda kullanmak için iyi bir optimizasyon algoritmasıdır. Adam algoritması, eğitim boyunca öğrenme oranını ayarlar.
    </li>
    <li>
        <b>binary_crossentropy</b> kayıp fonksiyonu bir örneğin iki sınıftan birisine ait olarak sınıflandırıldığı problemlerde kullanılmaktadır. Örneğin bir görüntünün kedi veya köpek olduğunun belirlenmesi gerekiyorsa devreye bu kayıp fonksiyonu girer. Örneğin köpek ise 1, kedi ise 0 olarak işaretlenebilir.
    </li>
    <li>
        Eğitim sırasında modelin nasıl bir performans gösterdiğini yorumlamak için her bir epoch sonunda validation seti ile elde edilen doğruluk ve loss miktarını görmek için <b>“accuracy”</b> metriği kullanıldı.
    </li>
</ul>


```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

<p></p>
Model, <b>150 epochs</b> değeriyle validasyon datası X_test, y_test olmak üzere eğitilmektedir.

```python
r = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=150)
```

Tüm derin öğrenme modellerini eğitirken kaydedilen varsayılan geri aramalardan biri <b>history()</b> aramadır. Her dönem için eğitim ölçümlerini kaydeder. Bu, eğer ayarlanmışsa, doğrulama veri kümesinin kaybı ve doğruluğunun yanı sıra(sınıflandırma sorunları için) kaybı ve doğruluğunu içerir.

<b>history</b> nesnesi, modeli eğitmek için kullanılan fit() işlevine yapılan çağrılardan döndürülür. Metrikler, döndürülen nesnenin geçmiş üyesindeki bir sözlükte saklanır.


```python
model.history.history
```

```python
import matplotlib.pyplot as plt
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss
```

```python
model_loss.plot()
```

<p></p>
<b>predictions</b> üzerinde tahminler oluşmaktadır. 

```python
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

<p></p>
<b>Sınıflandırma raporu</b> görselleştiricisi, model için kesinlik(precision), geri çağırma(recall), F1(fı-score) ve destek(support) puanlarını görüntüler. Rapor, daha kolay yorumlamayı ve sorun tespitini desteklemek için sayısal puanları renk kodlu bir ısı haritası ile birleştirir. Farklı sınıflandırma raporları arasında sınıflandırma modellerinin kolay karşılaştırmasını kolaylaştırmak için tüm ısı haritaları (0.0, 1.0) aralığındadır.

<ul>
    <li>
        <b>precision</b> bu sınıf arasında kaçının doğru şekilde sınıflandırıldığını gösterir.
    </li>
    <li>
        <b>recall</b> bu sınıftaki tüm eleman sayısı boyunca bu sınıftan kaç tane bulunduğunu gösterir.
    </li>
    <li>
        <b>fı-score</b> hassasiyet ve geri çağırma arasındaki harmonik ortalamadır.
    </li>
    <li>
        <b>support</b> belirli bir sınıfın veri kümesindeki oluşum sayısıdır. 
    </li>
</ul>

```python
print(classification_report(y_test,predictions))
```

<p></p>

<b>Karmaşıklık matrisi</b>, sınıflandırma problemi için bir performans ölçümüdür. Öngörülen ve gerçek değerlerin dört farklı kombinasyonunu içeren bir tablodur.Her ne olursa olsun sınıflandırma tahminleri şu dört değerlendirmeden birine sahip olacaktır:

<ul>
    <li>
        <b>TN(True Negative):</b> Hasta olmayana hasta değil demek.
    </li>
    <li>
        <b>TP(True Positive):</b> Hastaya hasta demek.
    </li>
    <li>
        <b>FN(False Negative):</b> Hasta olana hasta değil demek.
    </li>
    <li>
        <b>FP(False Positive):</b> Hasta olmayana hasta demek.
    </li>
</ul>

<b>Doğruluk(Accuracy)</b>, doğru olarak sınıflandırılan örneklerin yüzdesidir.
<p>accuracy = (TP+TN)/(TP+FP+TN+FN)</p>

<b>Kesinlik(precision)</b>, pozitif olarak tahminlenen değerlerin gerçekten kaç adedinin pozitif olduğunu göstermektedir.
<p>precision = (TP)/(TP+FP)</p>

<b>Duyarlılık(recall)</b>, pozitif olarak tahmin edilmesi gereken işlemlerin ne kadarını pozitif olarak tahmin edildiğini gösteren bir metriktir. “Gerçek pozitiflerin ne kadarı doğru bir şekilde tanımlandı?” sorusunu cevaplar.
<p>precision = (TP)/(TP+FN)</p>



```python
print(confusion_matrix(y_test,predictions))
```
